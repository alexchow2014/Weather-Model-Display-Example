```python
#@title Get rain data
def ECWMFRainData(productDate, productTime, forcastStart, forcastEnd):
  client = Client()

  client.retrieve(
      date=productDate,
      time=productTime,
      step=forcastStart,
      stream='oper',
      type="fc",
      param="tp",
      target="data.grib2",
  )

  client.retrieve(
      date=productDate,
      time=productTime,
      step=forcastEnd,
      stream='oper',
      type="fc",
      param="tp",
      target="dataPlus.grib2",
  )
```

```python
#@title grib2 to nc
def grib2ToNc(filename):
  grb = pg.open(filename)
  ncname = filename.replace('.grib2','.nc')
  da = nc.Dataset(ncname,'w',format='NETCDF4')

  g = grb[1]
  lat,lon = g.latlons()
  da.createDimension('lat',np.size(lat,0))
  da.createDimension('lon',np.size(lat,1))

  # add lat and lon variables
  da.createVariable('lat','f4',('lat','lon'),zlib=True,complevel=9,fill_value=-9999)
  da.createVariable('lon','f4',('lat','lon'),zlib=True,complevel=9,fill_value=-9999)
  da.variables['lat'][:]=lat
  da.variables['lon'][:]=lon

  for g in grb:

      # extract data and attribute
      varname = g.name
      data = g.values
      unit = g.units
      level = g.level

      # write into nc file
      if varname=='unknown':
          continue
      if level!=0:
          varname = varname + ' ' + str(level) + 'hPa'
      # print(varname)
      d = da.createVariable(varname,'f4',('lat','lon'),zlib=True,complevel=9,fill_value=-9999)
      d.units = unit
      da.variables[varname][:] = data

  da.close()
  grb.close()
  ncname
  return ncname
```

```python
#@title grib2 to opendataset

def grib2ToOpendataset():
  # ds is the "Total precipitation" at timePoint
  # dsPlus is the "Total precipitation" at timePoint+range
  ds = xr.open_dataset(grib2ToNc("data.grib2"))
  dsPlus = xr.open_dataset(grib2ToNc("dataPlus.grib2"))

  data = pygrib.open("data.grib2")
  g = data[1]
  lat, lon = g.latlons()

  # Limit the area of data we needed, here we got the data with in the area bounded by (26N, 108E), (26N, 120E), (18N, 120E) and (18N, 108E)
  ds = ds.where(lat >= 18).where(lat <= 26).where(lon >= 108).where(lon <= 120)
  dsPlus = dsPlus.where(lat >= 18).where(lat <= 26).where(lon >= 108).where(lon <= 120)

  # Find the difference of Total precipitation between dsPlus and ds
  PeriodOfRain = dsPlus["Total precipitation"] - ds["Total precipitation"]

  return PeriodOfRain
```

```python
#@title plot to dataset
def plot_dataset(dataset : xr.Dataset):
    projection = ccrs.Mercator()
    crs = ccrs.PlateCarree()

    plt.figure(figsize=(16,9), dpi=150)
    ax = plt.axes(projection=projection, frameon=False)

    import cartopy.feature as cf
    ax.add_feature(cf.COASTLINE.with_scale("10m"), lw=0.75)
    ax.add_feature(cf.BORDERS.with_scale("10m"), lw=0.3)

    lon_min = 108
    lon_max = 120
    lat_min = 18
    lat_max = 26

    cbar_kwargs = {'orientation':'horizontal', 'shrink':0.75, "pad" : .05, 'aspect':8, 'label':''}

    dataset.plot.contourf(ax=ax, transform=ccrs.PlateCarree(), levels=6+20, vmin=0, vmax=25)

    ax.set_extent([lon_min, lon_max, lat_min, lat_max], crs=crs)
    #plt.title(f"Temperature anomaly over Europe in")
    plt.show()
```

```python
#@title RainInformation function
def RainInformation(productDate, productTime, forcastStart, forcastEnd):

    # Get the data
    ECWMFRainData(productDate, productTime, forcastStart, forcastEnd)
    # Fatch the data
    PeriodOfRain = grib2ToOpendataset()

    # Change the Total precipitation provided by ECWMF from m to mm
    PeriodOfRain = np.fabs(PeriodOfRain) * 1000

    # Plot graph
    plot_dataset(PeriodOfRain)
```

```python
#@title Main function

# Variable: Date of forcast start
forcastDate = "2025-03-04"

# Variable: Time of forcast start, only 0, 6, 12 and 18 available, corresponding to 00UTC, 06UTC, 12UTC and 18UTC
forcastTime = 0

# Variable: starting forcast step point in hours, 0 to 144 (for 00UTC and 12UTC) or 0 to 90 (for 06UTC and 18UTC)
timePoint = 96+0

# Variable: Time range from starting forcast step point (in hours)
# If you would like to get total precipitation within 24 hours starting from starting forcast step point (in hours), enter 24
range = 24 #

# return image that the total precipitation within the period from starting forcast step point
# unit: mm
RainInformation(forcastDate, forcastTime, timePoint, timePoint+range)
```
